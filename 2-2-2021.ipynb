```python
import matplotlib.pyplot as plt
name =["a","b","c","d"]
marks=[344,347,233,345]
att=[224,350,300,25]
plt.plot(name,marks)
plt.plot(name,att)
#lt.show()
```

```python
plt.plot(name,marks,"og")
plt.plot(name,att,"*r")
plt.title("Student details")
plt.xlabel("names")
plt.ylabel("details")
plt.grid()
plt.legend()
```

```python
plt.plot(name,marks,label="marks")
plt.plot(name,att,label="att")
plt.title("Student details")
plt.xlabel("names")
plt.ylabel("details")
plt.grid()
plt.legend()
```

```python
plt.plot(name,marks,"--r",name,att,"og")
plt.ylim(200,400)
```

```python
plt.scatter(name,marks,color="grcm")
plt.scatter(name,att,color="grcm")
```

```python
plt.bar(name,att,color="grcm")
```

```python
plt.subplot(412)
plt.plot(name,marks,"om")
plt.subplot(431)
plt.plot(name,att,"*r")
plt.subplot(444)
plt.plot(att,marks,"--g")
plt.subplot(413)
plt.plot(att,name,"-k")
```
